Import Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

Load the Data

```python
data=pd.read_csv(r"C:\Users\hp\Desktop\placement.csv")
```

Exploring the Data

```python
data.shape
```

```python
data.describe()
```

Cleaning the Data

```python
data.isnull().sum()
```

Visualising the Data to find whether it is a linear relationship or not

```python
plt.figure(figsize=(16,6))
sns.scatterplot(data=data,x="cgpa",y="package")
plt.show()
```

Dividing the data into input and output

```python
x=data[["cgpa"]]
y=data["package"]
```

Dividing the data into training and testing data

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=32)
```

Linear Regression

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

Evaluating the Model

```python
lr.score(x_test,y_test)
```

```python
y_pred=lr.predict(x)
```

```python
Visualizing the results
```

```python
plt.figure(figsize=(16,6))
sns.scatterplot(data=data,x="cgpa",y="package")
plt.plot(data["cgpa"],y_pred,c="red")
plt.legend(["Original","Predict"])
plt.savefig("Predict.jpg")
plt.show()
```

```python
from sklearn.metrics import root_mean_squared_error
```

```python
root_mean_squared_error(y_test,lr.predict(x_test))
```
